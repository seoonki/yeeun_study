## 정규식을 테스트해 보자.

```
import re

pattern = r'best'
text = 'A best friend is one who walks in when the rest of the world walks out.'
match = re.search(pattern, text)
```

```
print(match)
```

```
if match :
  print(match)
  print(match.group())
  print(match.start())
  print(match.end())
  print(match.span())
```

```
pattern = '?est'
text = 'A best friend is one who walks in when the rest of the world walks out.'
match = re.search(pattern, text)
```

```
print(match)
```

## 정규표현식의 group() 기능을 테스트 하는 코드

```
import re
 
text = "Please call 010-2345-1234."
 
# group() 기능을 테스트하는 코드, 2개의 그룹이 생성된다
regex = re.compile('(\d{3})-(\d{4}-\d{4})')

match_obj = regex.search(text)
```

```
print(match_obj.group())
```

```
# 1 그룹
print(match_obj.group(1))
```

```
# 2 그룹
print(match_obj.group(2))
```

## 넘파이 로그 스케일 시각화

```
import numpy as np
import matplotlib.pyplot as plt 


x = np.linspace(0, 10**5, 10)  # y축은 선형 스케일
y = np.logspace(0, 5, 10)      # x축은 로그 스케일로 1에서 10**5까지의 값이 생성됨

# 선 그래프를 그린다. x축에는 years값, y축에는 gdp 값이 표시된다.
plt.plot(x, y, color='green', marker='o', linestyle='solid') 
 
# 제목을 설정한다. 
plt.title("lin scale vs log scale") 
 
# y축에 레이블을 붙인다. 
plt.show() 
```

## BMI와 target과의 상관관계 알아보자

```
import numpy as np
from sklearn import datasets 

# 당뇨병 데이터 세트를 sklearn의 데이터집합으로부터 읽어들인다. 
diabetes = datasets.load_diabetes()

# bmi 값을 1차원 배열로 추출함
X = diabetes.data[:, 2]
X
np.corrcoef(X, diabetes.target)
```

```
 regr = linear_model.LinearRegression() 

regr.fit(X, diabetes.target)         # 학습을 통한 선형회귀 모델을 생성 

# 학습 데이터와 테스트 데이터를 분리한다. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes.data[:,0:10],
                                                    diabetes.target,
                                                    test_size=0.2) 
regr = LinearRegression() 
regr.fit(X_train, y_train)

score = regr.score(X_train, y_train)
print(score)
score = regr.score(X_test, y_test)
print(score)
```
